```python
import pandas as pd
import os

# Input Excel file
excel_file = "Test_Data_OG - Copy.xlsx"

# Output folder for CSVs
output_dir = "exported_csvs"
os.makedirs(output_dir, exist_ok=True)

# Load all sheet names
xls = pd.ExcelFile(excel_file)

# Loop through sheets and save as CSV
for sheet in xls.sheet_names:
    df = pd.read_excel(excel_file, sheet_name=sheet)
    
    # Clean sheet name for filename
    clean_name = sheet.replace(" ", "_").replace("/", "_")
    
    # Export to CSV
    csv_path = os.path.join(output_dir, f"{clean_name}.csv")
    df.to_csv(csv_path, index=False, encoding="utf-8-sig")
    
    print(f"Exported {sheet} → {csv_path}")

print(" All sheets exported as CSVs")
```
