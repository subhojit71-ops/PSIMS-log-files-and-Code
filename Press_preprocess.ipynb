#Press_clean creation

```python
import csv

input_file = r'C:\ProgramData\MySQL\MySQL Server 8.0\Uploads\Press.csv'
output_file = r'C:\ProgramData\MySQL\MySQL Server 8.0\Uploads\Press_clean.csv'

with open(input_file, 'r', encoding='latin1') as infile, \
     open(output_file, 'w', newline='', encoding='latin1') as outfile:

    reader = csv.reader(infile, delimiter='\t')
    writer = csv.writer(outfile, delimiter='\t', quoting=csv.QUOTE_MINIMAL)

    # Copy header
    header = next(reader)
    writer.writerow(header)

    for row in reader:
        # Strip newlines/tabs from each field to prevent MySQL truncation errors
        clean_row = [field.replace('\n', ' ').replace('\r', ' ').replace('\t', ' ') for field in row]
        writer.writerow(clean_row)
```

```python
import csv

with open(r'C:\ProgramData\MySQL\MySQL Server 8.0\Uploads\Press.csv', 'r', encoding='utf-8', errors='ignore') as infile, \
     open(r'C:\ProgramData\MySQL\MySQL Server 8.0\Uploads\Press_clean.csv', 'w', newline='', encoding='utf-8') as outfile:

    reader = csv.reader(infile, delimiter='\t')
    writer = csv.writer(outfile, delimiter='\t', quoting=csv.QUOTE_ALL)

    header = next(reader)
    writer.writerow(header)

    for row in reader:
        clean_row = [cell.replace('\t', ' ').replace('\n', ' ').replace('\r', ' ') for cell in row]
        writer.writerow(clean_row)
```
